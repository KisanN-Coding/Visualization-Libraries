### **2D Line Plot**

**A Line Plot (also called a line chart) is a type of graph that connects individual data points with straight lines. It is commonly used to visualize trends or changes in values over time or across a continuous range**

* **Used For : Bivariate Analysis**

* **For Features : 1) Numerical --> Categorical Features and Numerical --> Numerical Features**


**Use a line plot when:**

- You want to see trends or patterns over time

- You are working with continuous data (data that changes gradually)

- You want to compare multiple series on the same plot

- You are analyzing relationships between two variables (e.g., x and y)

```python
rate = [10000,14000,16000,20000,18000,13000,9000,15000]

year = [2020,2021,2022,2023,2024,2025,2026,2027]

plt.plot(year,rate)
```

**1 ) color parameter**

In Matplotlib, the color parameter in plt.plot() (and other plotting functions) is used to set the color of the line or markers in the graph

```python
plt.plot(data['index'],data['RG Sharma'],color='c')
plt.plot(data['index'],data['V Kohli'],color='g')

plt.title('Runs Analysis VK vs RS')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

```python
# Applying the colors using hex code (Using hex code we can apply any color to our graph)

plt.plot(data['index'],data['RG Sharma'],color='#269E76')
plt.plot(data['index'],data['V Kohli'],color='#BAA427')

plt.title('Runs Analysis VK vs RS')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

**Changing the style of line**

In Matplotlib, the linestyle parameter in plt.plot() controls how the line looks â€” solid, dashed, dotted, etc

```python
plt.plot(data['index'],data['RG Sharma'],color="#39CA99",linestyle='--')
plt.plot(data['index'],data['V Kohli'],color="#A8D016",linestyle='--')

plt.title('Runs Analysis VK vs RS')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

**3 ) linewidth parameter**

In Matplotlib, the linewidth (or shorter lw) parameter in plt.plot() sets how thick the line will appear

```python
plt.plot(data['index'],data['RG Sharma'],color="#2CBC5C",linewidth=2)
plt.plot(data['index'],data['V Kohli'],color='#BAA427',linewidth=5)

plt.title('Runs Analysis VK vs RS')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

**4 ) marker parameter**

In Matplotlib, the marker parameter in plt.plot() controls the symbol used to mark each data point on the line.
Itâ€™s useful when you want to highlight individual points instead of just the line

```python
plt.plot(data['index'],data['RG Sharma'],color="#2CBC5C",marker='o')
plt.plot(data['index'],data['V Kohli'],color='#BAA427',marker='o')

plt.title('Runs Analysis VK vs RS')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')

# We can use many more marker styles like , < , > , ^ , * , etc , you can try them
```

**5 ) markersize**

There is marker size parameter to manipulate the size of markers

```python
plt.plot(data['index'],data['RG Sharma'],color="#2CBC5C",marker='o',markersize=15)
plt.plot(data['index'],data['V Kohli'],color='#BAA427',marker='o',markersize=15)

plt.title('Runs Analysis VK vs RS')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

**6 ) label parameter and legend function**

In Matplotlib, the label parameter is used to give a name to a plotted line or dataset, and the legend() function displays those names in a legend box on the plot

```python
plt.plot(data['index'],data['RG Sharma'],color="#2CBC5C",label='RS')
plt.plot(data['index'],data['V Kohli'],color='#BAA427',label='VK')

plt.legend() # the legend function

plt.title('Runs Analysis VK vs RS')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

```python
# Using the loc parameter , we can also change the location of legend box (By default is 'best')

plt.plot(data['index'],data['RG Sharma'],color="#2CBC5C",label='RS')
plt.plot(data['index'],data['V Kohli'],color='#BAA427',label='VK')

plt.legend(loc='center left')

plt.title('Runs Analysis VK vs RS')
plt.xlabel('Seasons')
plt.ylabel('Runs Scored')
```

### **Scatter Plots**

**A scatter plot is a type of graph used to display the relationship between two numerical variables.**
**Each point on the plot represents one observation from your data, with:**

- X-axis â†’ the value of one variable

- Y-axis â†’ the value of the other variable

**Itâ€™s especially useful for spotting patterns, trends, clusters, and outliers in data**

**Key Points :**

- **Purpose :** Show correlation or relationship between two variables

- **Data type :** Both variables should be numerical

- **Appearance :** A collection of points scattered on the plot area

- **Extras :** You can change color, size, and style to add more meaning

```python
x = np.linspace(-10,10,50)
y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
# ploting the scatter plot

plt.scatter(x,y)
```

## **Bar Chart**

**A bar chart in Matplotlib is a way to represent categorical data (like names, categories, or groups) with rectangular bars whose heights (or lengths, in horizontal form) correspond to the values they represent**

**When to Use a Bar Chart :**

- Compare different categories

- Show ranking (highest to lowest)

- Display small discrete time comparisons

- Show composition (stacked bars)

- Compare groups side-by-side (grouped bars)

```python
# Simple bar chart

children = [10,30,60,40,20,80]
foods = ['burger','icecream','chocolate','pizza','candy','milkshake']
```

```python
plt.bar(foods,children)
```

**For horizontal bars**

```python
plt.barh(foods,children,color='g')  # More useful when number of categories increases (eg. 8-10 or more )
```

```python
batsmans = pd.read_csv('Datasets/batsman_season_record.csv')
batsmans
```

#### **Multiple Bar Charts**

```python
np.arange(batsmans.shape[0])
```

```python
plt.bar(np.arange(batsmans.shape[0]) - 0.2,batsmans['2015'],width=0.2)
plt.bar(np.arange(batsmans.shape[0]),batsmans['2016'],width=0.2)
plt.bar(np.arange(batsmans.shape[0]) + 0.2,batsmans['2017'],width=0.2)

plt.xticks(np.arange(batsmans.shape[0]),batsmans['batsman'],rotation=30)  # xticks function to change the labels
plt.show()
```

```python
# To do that in one line  (Using the plot function)

batsmans.set_index('batsman')[['2015','2016','2017']].plot(kind='bar')
```

#### **A Problem with labels**

```python
children = [10,30,60,40,20,80]
foods = ['burgerrrrr','icecreammmmmmm','chocolateeee','pizzaaaaa','candyyyyy','milkshakeeee']

plt.bar(foods,children)  # See the labels on x axis
```

**So, we can you xticks with rotation parameter**

```python
plt.bar(foods,children)
plt.xticks(rotation=60)
plt.show()
```

#### **Stacked Bar Chart**

```python
plt.bar(batsmans['batsman'],batsmans['2017'],bottom = (batsmans['2015'] + batsmans['2016']),label='2017')
plt.bar(batsmans['batsman'],batsmans['2016'],bottom=batsmans['2015'],label='2016')
plt.bar(batsmans['batsman'],batsmans['2015'],label='2015')

plt.legend()
plt.xticks(rotation=30)
plt.show()
```

### **Pie Chart**
