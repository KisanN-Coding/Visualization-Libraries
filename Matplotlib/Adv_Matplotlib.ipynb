### **Colored Scatterplot**

```python
# Loading the iris dataset

iris = pd.read_csv('Datasets/iris.csv')
iris.sample(5)
```

```python
# Converting iris['Species'] labels to 0 ,1 ,2

iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}).infer_objects(copy=False)
iris.sample(5)

# .infer_objects(copy=False) # This is just to remove the warning
```

##### **1 ) c parameter**

**In plt.scatter(), the c parameter controls the color of each point in the scatter plot**

##### **2 ) cmap parameter**

**In Matplotlib, the cmap parameter stands for colormap — a predefined mapping from numbers to colors**

**It’s used when your c parameter (color input) is numeric rather than a fixed color name**

```python
cs = plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

# Ignore , below code for now 

cbar = plt.colorbar(cs,ticks=[0,1,2])
cbar.ax.set_yticklabels(['setosa', 'versicolor', 'virginica'])
plt.show()
```

##### **3 ) alpha parameter**

**In Matplotlib, the alpha parameter controls the transparency (opacity) of the plotted elements**

```python
cs = plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

# Ignore , below code for now 

cbar = plt.colorbar(cs,ticks=[0,1,2])
cbar.ax.set_yticklabels(['setosa', 'versicolor', 'virginica'])
plt.show()
```

#### **Plot Size**

```python
plt.figure(figsize=(8,6))  # Code line to be focused

cs = plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

# Ignore , below code for now 

cbar = plt.colorbar(cs,ticks=[0,1,2])
cbar.ax.set_yticklabels(['setosa', 'versicolor', 'virginica'])
plt.show()
```

### **Annotations**

**In Matplotlib, annotations are extra pieces of text, arrows, or markers you add to a plot to highlight or explain specific data points, trends, or region**

```python
batter = pd.read_csv('Datasets/batter.csv')
batters = batter.head(25)
batters
```

```python
plt.figure(figsize=(15,10))
plt.scatter(batters['avg'],batters['strike_rate'],s=batters['runs'])

for i in range(batters.shape[0]):
    plt.text(batters['avg'].values[i],batters['strike_rate'].values[i],batters['batter'].values[i])
```

#### **Horizontal and Vertical Lines**

```python
plt.figure(figsize=(15,10))
plt.scatter(batters['avg'],batters['strike_rate'],s=batters['runs'])

plt.axhline(135,color='g') # axis horizontal line 
plt.axvline(38,color='g')

for i in range(batters.shape[0]):
    plt.text(batters['avg'].values[i],batters['strike_rate'].values[i],batters['batter'].values[i])
```

### **Different way to plot the graphs**

```python
batters.head()
```

```python
# Normal way

plt.scatter(batters['avg'],batters['strike_rate'])
```

```python
# Another way # Advantage of this is, we can plot multiple plots side by side

fig , ax = plt.subplots(figsize=(5,5))

ax.scatter(batters['avg'],batters['strike_rate'])

ax.set_title('Average vs Strike Rate') # For title of plot

ax.set_xlabel('Average')  # For x label
ax.set_ylabel('Strike Rate') # For y label

plt.show()
```

### **Subplots**

```python
fig, ax = plt.subplots(nrows=2,ncols=1,figsize=(6,10),sharex=True) # Sharex make x axis common for both

ax[0].scatter(batter['avg'],batter['strike_rate'],color="#6AB811")
ax[0].set_title('Avg vs Strike rate')
ax[0].set_xlabel('Avg')
ax[0].set_ylabel('Strike Rate')

ax[1].scatter(batter['avg'],batter['runs'],color="#B5580C")
ax[1].set_title('Avg vs Runs')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batter['avg'],batter['strike_rate'],color="#6AB811")
ax[0,0].set_title('Avg vs Strike rate')
ax[0,0].set_xlabel('Avg')
ax[0,0].set_ylabel('Strike Rate')

ax[0,1].scatter(batter['avg'],batter['runs'],color="#B5580C")
ax[0,1].set_title('Avg vs Runs')
ax[0,1].set_xlabel('Avg')
ax[0,1].set_ylabel('Runs')

ax[1,0].hist(batter['avg'],color="#6AB811")

ax[1,1].hist(batter['avg'],color="#B5580C")

plt.show()
```

#### **3rd Way to plot graph**

```python
plt.figure(figsize=(10,5))

ax1 = plt.subplot(1,2,1)
ax1.scatter(batter['avg'],batter['strike_rate'],color="#A82EACFF")

ax2 = plt.subplot(1,2,2)
ax2.hist(batter['avg'],color="#A82EACFF")

plt.show()
```

### **3D Plots**

#### **1 ) 3D Scatter Plot**

#### **2 ) 3D Line Plot**

```python
x = [0,1,5,25]
y = [0,10,13,9]
z = [0,13,20,5]

fig = plt.figure()

ax = fig.add_subplot(projection='3d')
ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='r')
```

#### **3 ) 3D Surface Plot**

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
```

```python
fig = plt.figure()

ax = fig.add_subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='plasma')
fig.colorbar(p)
```

```python
zz = np.cos(xx) + np.sin(yy) + np.cos(xx)
fig = plt.figure()

ax = fig.add_subplot(projection='3d')
p = ax.plot_surface(xx,yy,zz,cmap='plasma')
fig.colorbar(p)
```

#### **4 ) Contour Plot**

```python
fig = plt.figure()

ax = fig.add_subplot()
p = ax.contour(xx,yy,z,cmap='plasma')
fig.colorbar(p)
```

```python
# Projection of 3D Graph in 2D

# 3D 

fig = plt.figure(figsize=(12,4))

ax = fig.add_subplot(1,2,1,projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='plasma')

# 2D Contour ( a filled version)

ax1 = fig.add_subplot(1,2,2)
p = ax1.contourf(xx,yy,z,cmap='plasma')
fig.colorbar(p)

```

```python
fig = plt.figure()

ax = fig.add_subplot()
p = ax.contour(xx,yy,zz,cmap='viridis')
fig.colorbar(p)
```

```python
# Projection of 3D Graph in 2D

# 3D 

fig = plt.figure(figsize=(12,4))

ax = fig.add_subplot(1,2,1,projection='3d')
p = ax.plot_surface(xx,yy,zz,cmap='viridis')

# 2D Contour ( a filled version)

ax1 = fig.add_subplot(1,2,2)
p = ax1.contourf(xx,yy,zz,cmap='viridis')
fig.colorbar(p)
```

#### **Pie Chart**

```python
df = pd.DataFrame(
    {
        'batsman':['Dhawan','Rohit','Kohli','SKY','Pandya','Pant'],
        'match1':[120,90,35,45,12,10],
        'match2':[0,1,123,130,34,45],
        'match3':[50,24,145,45,10,90]
    }
)

df.head()
```

```python
df['match1'].plot(kind='pie',labels=df['batsman'].values,autopct='%0.1f%%')
```

#### **subplots parameter**

```python
df[['match1','match2','match3']].set_index(df['batsman']).plot(kind='pie',subplots=True,autopct='%0.1f%%',figsize=(25,10))
plt.show()
```
