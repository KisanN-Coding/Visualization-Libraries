## **Seaborn**

**Seaborn is a Python library for data visualization, built on top of Matplotlib. It makes it easier to create beautiful, statistical plots with very little code**

**Key Features :**

- Statistical plots: Specially designed for distributions, correlations, regressions, heatmaps, boxplots, violin plots, etc.

- Better style & colors: Comes with attractive default themes and color palettes.

- Works with Pandas: Integrates seamlessly with Pandas DataFrames, so you can pass data directly.

- Less code, more visuals: What may take many lines in Matplotlib can often be done in 12 lines with Seaborn.

**Seaborn = A prettier, easier, and statistics-friendly version of Matplotlib**




**Types of functions in seaborn :**

1 ) Figure Level

2 ) Axis Level

```python
# Required Libraries

import seaborn as sns
import matplotlib.pyplot as plt

# Other 

import pandas as pd 
import numpy as np
import plotly.express as px
```

```python
# Style

plt.style.use('fivethirtyeight')
```

### **1. Relational Plot**

- To see the statistical relation between 2 or more variables

- Used for Bivariate Analysis

**Plots under this section**

- Scatterplot

- Lineplot

#### **1 ) Scatterplot**

```python
# Loading the dataset

tips = sns.load_dataset('tips')
tips.sample(5)
```

**Axis Level Function**

```python
# scatterplot

sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# hue parameter  separates by color

sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# style parameter  separates by marker shape/line dash

sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker')
```

```python
# size parameter  separates by marker size/line thickness

sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker',size='size')
```

**Figure Level Function**

```python
# relplot

sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='smoker',size='size')

# If you not provide kind , by default it is scatter
```

#### **2 ) Line plot**

```python
# Loading the data

gap = px.data.gapminder()
gap.sample(5)
```

```python
temp_df = gap[gap['country'] == 'India']
temp_df.sample(5)
```

**Axis Level Function**

```python
# Line plot

sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

**Figure Level Function**

```python
# relplot

sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
# All parameters ---> hue , style , size

temp_df = gap[gap['country'].isin(['Brazil','India','Iceland'])]
temp_df.sample(5)
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',hue='country',style='continent',kind='line',size='iso_alpha')
```

#### **Facet plot Concept**

A Facet Plot means splitting one big plot into multiple smaller subplots (called "facets"), based on the values of one or more categorical variables

**Note: This will only work with figure level function**

```python
# col (column) parameter

sns.relplot(data=tips,x='total_bill',y='tip',col='sex')
```

```python
# row parameter

sns.relplot(data=tips,x='total_bill',y='tip',row='sex')
```

```python
# Both parameter together

sns.relplot(data=tips,x='total_bill',y='tip',row='sex',col='time')
```

```python
# Col wrap parameter 

sns.relplot(data=gap,x='lifeExp',y='gdpPercap',col='year')
```

```python
# The above plot is not useful , so need to wrap up the column (For example 3,4 in a row)
# To do this we use col_wrap


sns.relplot(data=gap,x='lifeExp',y='gdpPercap',col='year',col_wrap=3)  # 3 col in row
```

### **2. Distribution Plots**

- Used for univariate analysis

- Used to find out the distribution

- Range of the observation

- Get to know Central Tendency

- Is the data bimodal?

- Are there outliers?

**Plots under distribution plot :**

- Histplot

- Kdeplot

- Rugplot

#### **1 ) Hist plot**

**Axis Level Function**

```python
# hist plot

sns.histplot(data=tips,x='total_bill')
```

**Figure Level Function**

```python
sns.displot(data=tips,x='total_bill',kind='hist')

# If is not provide the kind parameter, by default it is hist
```

```python
# bins parameter

sns.displot(data=tips,x='total_bill',bins=20,kind='hist')
```

```python
# hue parameter

sns.displot(data=tips,x='total_bill',kind='hist',hue='sex')
```

```python
# element parameter

sns.displot(data=tips,x='total_bill',kind='hist',hue='sex',element='step')
```

```python
# Loading Titanic dataset

titanic = sns.load_dataset('titanic')
titanic.sample(5)
```

```python
sns.displot(data=titanic,x='age',kind='hist',element='step')
```

```python
# Facet plot 

sns.displot(data=titanic,x='age',kind='hist',element='step',col='sex')
```

#### **2 ) Kde plot**

**It is a smoothed version of a histogram that shows the probability density (distribution) of a continuous variable. Instead of showing bars like a histogram, it draws a smooth curve that represents where the data points are concentrated**

**Axis Level Function**

```python
# kdeplot

sns.kdeplot(data=tips,x='total_bill')
```

**Figure Level Function**

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue parameter

sns.displot(data=tips,x='total_bill',kind='kde',hue='sex')
```

```python
# fill parameter 

sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

#### **2 ) Rug plot**

**A rug plot is a simple plot that shows each data point as a small tick mark on an axis, like laying a rug of lines along it**
**It helps you quickly see where data points are dense or sparse**

```python
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',fill=True,rug=True)
```

#### **Bivariate Histogram**

```python
sns.displot(data=tips,x='total_bill',y='tip')
```

#### **Bivariate Kdeplot**

```python
sns.displot(data=tips,x='total_bill',y='tip',kind='kde')
```

### **Changing the size of figure level function**

```python
# There are two parameters for it one is height and other

sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',height=10,aspect=2)
```

### **3. Matrix Plot**

**In Seaborn, matrix plots are plots that show data in a grid (matrix) form, where rows and columns represent categories or variables, and the color/shade of each cell shows the value**

- Heatmap

- Clustermap

#### **1 ) Heat Map**

**A heatmap is a type of matrix plot that shows data values as colors inside a grid**

```python
gap.sample(5)
```

```python
temp_df = gap.pivot_table(index='country',columns='year',values='lifeExp')
temp_df
```

```python
# Heatmap --> Axis Level Function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
tp_df = gap[gap['continent'] == 'Europe'].pivot_table(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(tp_df)
```

```python
# Parameters ---> annot , linewidth , cmap 
plt.figure(figsize=(15,15))
sns.heatmap(tp_df,annot=True,linewidths=0.6,cmap='plasma')
```

#### **2 ) Cluster Map**

**A ClusterMap is a special kind of heatmap in Seaborn that also performs clustering (grouping similar rows and columns together)**

- It looks like a heatmap

- But you also get dendrograms (tree diagrams) on the top and side, showing how rows/columns are grouped

- Similar rows/columns are moved next to each other automatically

```python
iris = px.data.iris()
iris.sample(5)
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
